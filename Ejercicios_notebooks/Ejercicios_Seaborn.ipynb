````python
sns.histplot(data=titanic, x="fare",element="step",color='r' )
plt.ylabel('')
````

````python
# EL CÓDIGO AQUÍ
# REPLICAR LA IMAGEN DE LA GRÁFICA CONTINUACIÓN
# TENGA CUIDADO DE NO SOBRESCRIBIR LA CELDA DE ABAJO
# ¡QUE QUITARÍA LA IMAGEN DE LA GRÁFICA DEL EJERCICIO!
sns.boxplot(x='class',y='age',data=titanic,hue='who')
````

````python
# EL CÓDIGO AQUÍ
# REPLICAR LA IMAGEN DE LA GRÁFICA CONTINUACIÓN
# TENGA CUIDADO DE NO SOBRESCRIBIR LA CELDA DE ABAJO
# ¡QUE QUITARÍA LA IMAGEN DE LA GRÁFICA DEL EJERCICIO!
sns.swarmplot(x='class',y='age',data=titanic)
````

````python
# EL CÓDIGO AQUÍ
# REPLICAR LA IMAGEN DE LA GRÁFICA CONTINUACIÓN
# TENGA CUIDADO DE NO SOBRESCRIBIR LA CELDA DE ABAJO
# ¡QUE QUITARÍA LA IMAGEN DE LA GRÁFICA DEL EJERCICIO!
import pandas  as  pd
frequency_sex= titanic['sex'].value_counts()
df_frequency_sex = pd.DataFrame({'sex': frequency_sex.index.tolist(), 'count': frequency_sex.tolist()})
sns.barplot(x='sex', y='count', data= df_frequency_sex )
````

````python
# EL CÓDIGO AQUÍ
# REPLICAR LA IMAGEN DE LA GRÁFICA CONTINUACIÓN
# TENGA CUIDADO DE NO SOBRESCRIBIR LA CELDA DE ABAJO
# ¡QUE QUITARÍA LA IMAGEN DE LA GRÁFICA DEL EJERCICIO!

sns.heatmap(titanic.corr(), vmin = -1, vmax = +1, cmap = 'coolwarm')
````
